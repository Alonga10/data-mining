### Support Vector Machine

```python
X = rng.standard_normal((200, 2))
X[:100] += 2
X[100:150] -= 2
y = np.array([1]*150+[2]*50)
```

```python
fig , ax = subplots(figsize=(8,8))
ax.scatter(X[:,0],
    X[:,1],
    c=y,
    cmap=cm.coolwarm)
```

```python
(X_train ,
X_test ,
y_train ,
y_test) = skm.train_test_split(X,
y,
test_size=0.5,
random_state=0)
svm_rbf = SVC(kernel="rbf", gamma=1, C=1)
svm_rbf.fit(X_train , y_train)
```

```python
fig , ax = subplots(figsize=(8,8))
plot_svm(X_train ,
    y_train ,
    svm_rbf ,
    ax=ax)
```

```python
svm_rbf = SVC(kernel="rbf", gamma=1, C=1e5)
svm_rbf.fit(X_train , y_train)
fig , ax = subplots(figsize=(8,8))
plot_svm(X_train ,
    y_train ,
    svm_rbf ,
    ax=ax)
```

```python
kfold = skm.KFold(5,
random_state=0,
shuffle=True)
grid = skm.GridSearchCV(svm_rbf ,
{'C':[0.1,1,10,100,1000],
'gamma':[0.5,1,2,3,4]},
refit=True ,
cv=kfold ,
scoring='accuracy');
grid.fit(X_train , y_train)
grid.best_params_
```

```python
best_svm = grid.best_estimator_
fig , ax = subplots(figsize=(8,8))
plot_svm(X_train ,
         y_train ,
        best_svm ,
        ax=ax)
y_hat_test = best_svm.predict(X_test)
confusion_table(y_hat_test , y_test)
```

### ROC Curves

```python
fig , ax = subplots(figsize=(8,8))
roc_curve(best_svm ,
    X_train ,
    y_train ,
    name='Training',
    color='r',
    ax=ax);
```

```python
svm_flex = SVC(kernel="rbf",
gamma=50,
               C=1)
svm_flex.fit(X_train , y_train)
fig , ax = subplots(figsize=(8,8))
roc_curve(svm_flex ,
    X_train ,
    y_train ,
    name=r'Training $\gamma=50$',
    color='r',
    ax=ax);
```

```python
roc_curve(svm_flex ,
    X_test ,
    y_test ,
    name=r'Test $\gamma=50$',
    color='b',
    ax=ax)
fig;
```

```python
fig , ax = subplots(figsize=(8,8))
for (X_, y_, c, name) in zip(
    (X_train , X_test),
    (y_train , y_test),
    ('r', 'b'),
    ('CV tuned on training',
    'CV tuned on test')):
    roc_curve(best_svm ,
        X_,
        y_,
        name=name ,
        ax=ax,
    color=c)
```

### SVM with Multiple Classes
